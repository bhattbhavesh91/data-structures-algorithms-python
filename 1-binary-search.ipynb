```python
def binary_search(inp_list, item):
    low = 0
    high = len(inp_list) - 1

    while low <= high:
        mid_index = (low + high)//2
        mid_value = inp_list[mid_index]

        if item == mid_value:
            return mid_index
        if mid_value > item:
            high = mid_index - 1
        else:
            low = mid_index + 1
    return None
```
